```python
import pandas as pd
```

```python
dataset=pd.read_csv(r'Restaurant_Reviews.tsv', sep='\t')
```

```python
all_article=dataset['Review'].values
```

```python
print(all_article.size)
```

```python
print(all_article)
```

```python
#計算有多少個句子是以 . 結尾
end_dot=0
for i in range(0,1000):
    if format(all_article[i]).endswith('.') == True:
        end_dot=end_dot+1
print('There are',end_dot,'sentences end with ".".')
```

```python
將所有. 換成 ,
print(format(all_article).replace('.',','))
```

```python
#將所有sentence 中的第一個 the 置換成 The
for i in range(0,1000):
    if format(all_article[i].find('the')) == -1:
        print(all_article[i])
    else:
        print(all_article[i].replace('the','The',1))
```

```python
#將偶數句子全部轉換為大寫，基數句子全部轉換為小寫
for i in range(0,1000):
    if i%2 ==1:
        print(all_article[i].upper())
    else:
        print(all_article[i])
```

```python
#將所有句子合併在一起，並以' / ' 為間隔
for i in range(0,1000):
    print(format(all_article[i]),end=' / ')
```
